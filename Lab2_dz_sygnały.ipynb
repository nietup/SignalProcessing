## Modulacja

### Modulacja AM

Modulacja amplitudowa jest prosta bo polega na zastąpienie parametru amplitudy sygnału nośnego, wartością sygnału modulującego. Weźmy dla przykłady 3 sekundowy sygnał o częstotliwości 30 Hz (Fs=1kHz). Zmoduluj jego amplitudę sygnałem o częstotliwości 2 Hz i narysuj pod spodem.

```python
def sins(A, f, p, l, Fs):
    s = 1/Fs

    x = arange(0, l, s)
    y = A*sin(2*pi*f*x + p)
    return x, y
```

```python
x, m = sins(5, 2, 0, 10, 1000)
_, y = sins(m, 300000, 0, 10, 1000)

display(Audio(y, rate=1000))
plo()
```

### Modulacja FM

Modulacja częstotliwościowa działa podobnie do modulacji amplitudowej, ale zamiast amplitudy zmienia częstotliwość sygnału nośnego. Biorąc te same parametry jak w poprzednim zadaniu (3s, 30Hz, Fs=1kHz) wykonaj modulację częstotliwościową.

Warto też wziąć pod uwagę, że rozrzut amplitudy sygnału nadawanego jest w dziedzinie częstotliowści, więc powinien wynosić nieco więcej, aby był słyszalny (conajmniej 10, 20 Hz). 

```python
T = 3
Fs = 8000

t = arange(0, T, 1/Fs)

A = 1
ph = 0
fc = 1000
fm = 2

ym = A*sin(2*pi*fm*t)
y = A*sin(2*pi*fc*t+3*ym)

_=specgram(y, Fs=Fs)
```

Wygeneruj sygnał wyrażony jako suma sygnałów o parametrach podanych poniżej. Sygnał ma mieć 2.5 sekundy, przy częstotliwości próbkowania 44.1 kHz. Narysuj i odtwórz taki dźwięk.

1. f=500Hz A=2 $\phi$=0 rad
2. f=1234Hz A=0.5 $\phi$=$\pi$ rad
3. f=100Hz A=1 $\phi$=115$^\circ$

```python
l = 2.5
Fs = 44100

x = arange(0, l, 1/Fs)
y1 = 2*sin(2*pi*500*x)
y2 = 0.5*sin(2*pi*1234*x + pi)
y3 = sin(2*pi*100*x + radians(115))

y = y1 + y2 + y3

figure(figsize=(10, 10))
subplot(2, 2, 1)
title("Wynikowy sygnał")
plot(x, y)

subplot(2, 2, 2)
title("Składowa f=500Hz A=2 ϕ=0 rad")
plot(x, y1)

subplot(2, 2, 3)
title("Składowa f=1234Hz A=0.5 ϕ=π rad")
plot(x, y2)

subplot(2, 2, 4)
title("Składowa f=100Hz A=1 ϕ=115∘")
plot(x, y3)

display(Audio(y, rate=Fs))
```

Z powyższego sygnału wylicz i narysuj jego składową parzystą i nieparzystą (patrz wykład). Potem dodaj te składowe do siebie tak żeby znowu otrzymać oryginalny sygnał. Do odwracania sewkecji użyj metody *np.flip*.

```python
yf = flip(y, 0)
yp = y + yf
yn = y - yf
ynew = 0.5*(yp + yn)

"""
odejmujemy stary sygnał od nowego żeby sprawdzić czy są takie same
przy obliczeniach powstaje pomijalny błąd (reprezentacji l. zmiennoprzecinkowych ?)
rzędu 4*1e-16
"""
ytest = ynew - y

plot(x, ytest)
```

Wygeneruj sekundę sygnału o częstotliwości próbkowania 10 kHz z parametrów poniżej i narysuj. Czemu uzyskałeś taki wykres (zwróć uwagę na zakres wartości)?

1. f=100 Hz A=1 $\phi$=$\frac{\pi}{2}$
2. f=200 Hz A=2 $\phi$=$\pi$
3. f=100 Hz A=-1 $\phi$=$\frac{\pi}{2}$
4. f=200 Hz A=2 $\phi$=$2\pi$

```python
l = 1
Fs = 10000

x = arange(0, l, 1/Fs)

y1 = sin(2*pi*100*x + 0.5*pi)
y2 = 2*sin(2*pi*200*x + pi)
y3 = -sin(2*pi*100*x + 0.5*pi)
y4 = 2*sin(2*pi*200*x + 2*pi)
y = y1 + y2 + y3 + y4

figure(figsize=(20, 10))
title("Wynikowy sygnał")
plot(x, y)

display(Audio(y, rate=Fs))
```

```python
y13 = y1 + y3

y24 = y2 + y4

figure(figsize=(20, 10))
subplot(1, 2, 1)
title("Składowe 1 i 3")
plot(x, y13)

subplot(1, 2, 2)
title("Składowe 2 i 4")
plot(x, y24)
```

### Omówienie

Składowe **1 i 3** oraz **2 i 4** parami się niwelują.

W przypadku pary **1 i 3** mamy idealnie wartość 0 w każdym punkcie ponieważ składowe te różnią się tylko amplitudami, które są liczbami odwrotnymi.

Dla pary **2 i 4** fazy to odpowiednio pół i cały okres sinusa,
$$sin(x + \pi) = -sin(x)$$
$$sin(x + 2\pi) = sin(x)$$
mamy tu jednak do czynienia z błędem reprezentacji liczb zmiennoprzecinkowych, w związku z czym suma tych składowych przyjmuje niezerowe wartości rzędu $10^{-13}$.
